```python
#supporting dataframes to show where we land and continue transforming
bigframe = pd.DataFrame(dict([(k,pd.Series(v)) for k,v in data['db'][0].items()]))
bigframe
```

```python
normbigframe = pd.json_normalize(json.loads(bigframe.to_json(orient='records')))
normbigframe
```

```python
normbigframe2 = pd.json_normalize(json.loads(normbigframe['data'].to_json(orient='index')))
normbigframe2
```

```python
normbigframe3 = pd.DataFrame(dict([(k,pd.Series(v)) for k,v in normbigframe2.items()])).T
normbigframe3
```

```python
#single post acquired
#normbigframe3[0][6][0] is depreciating so make sure to get used to iloc
normbigframe3.iloc[6,0][0]
#type(normbigframe3.iloc[6,0])
```

from post tags portion, I need post_id and tag_id so I can join later
for post_tags 
for tagging, join in tag_id and id

```python
#make sure I have the correct rows to break out
bigframe.iloc[[10,19]]
#posts_tags broken out
post_tags = bigframe.iloc[10,1]
post_tagsdf = pd.DataFrame.from_dict(post_tags, orient='columns')
#still need post_id for later join, 
post_tagsdf.drop(['id', 'sort_order'], axis = 1, inplace=True)
post_tagsdf
```

```python
#tags broken out
tags = bigframe.iloc[19,1]
tagsdf = pd.DataFrame.from_dict(tags, orient='columns')
#dropping literally everything else because don't need it
tagsdf.drop(tagsdf.columns.difference(['id', 'name', 'description']), axis=1, inplace=True)
tagsdf
```

```python
#left join on post_tagsdf['tag_id'] and tagsdf['id']
postTagsDF = pd.merge(post_tagsdf, tagsdf, left_on='tag_id', right_on='id', how='left')
postTagsDF
#now merged, going to drop the drop the now unnecessary and redundant columns
postTagsDF.drop(['tag_id', 'id'], axis=1, inplace= True)
postTagsDF
```

```python
#merge the last dfs to what I need them to be
completeDF = pd.merge(pd.DataFrame.from_dict(normbigframe3.iloc[6,0], orient='columns'), postTagsDF, left_on='id', right_on='post_id')
completeDF
#id, title, html, plaintext, published_at, name, description, 

completeDF.drop(completeDF.columns.difference(['id', 'title', 'published_at', 'html', 'plaintext', 'name', 'description']), axis=1, inplace=True)
completeDF
```

```python
completeDF['published_at'] = pd.to_datetime(completeDF['published_at'])
completeDF.sort_values(by='published_at', ascending=True, inplace=True)
```

```python
from bs4 import BeautifulSoup
def neat_function(i):
    print(f'{completeDF['title'][i]}')
    print(f'Originally published to www.fromthelemontrees.com on {completeDF['published_at'][i].strftime('%m.%d.%Y')}\n\n')
    print(BeautifulSoup(completeDF['plaintext'][i], 'html.parser').prettify())
    print(f'{completeDF['name'][i]}')
    print('\n\n')
neat_function(5)
```

```python
#loop through df column
publishing = []
for index, row in completeDF.iterrows():
    publishing.append(neat_function(index))
print(publishing)
```
